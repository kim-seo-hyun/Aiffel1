```python
# groupby()
# 데이터를 특정 기준으로 그룹핑할 때 활용. 엑셀의 피봇테이블과 유사
# 기준 컬럼을 중심으로 통계를 보는 것
# 반드시 aggregate하는 통계함수와 일반적으로 같이 적용

df.groupby('column').mean()
df.groupby(['column', 'column1']).mean() # 2개 이상의 컬름으로 그룹할 때 리스트로 묶어서 지정
df.groupby(['column', 'column1']).['column2'].mean() # column2의 결과만 도출하고 싶다면 단, series로 출력
df.groupby(['column', 'column1']).[['column2']].mean() # df로 출력하고 싶을 땐 []추가
pd.DataFrame(df.groupby(['column', 'column1']).['column2'].mean()) # DF로 출력

df.groupby(['column', 'column1']).['column2'].mean().reset_index() # 그룹핑된 데이터프레임의 인덱스 초기화
df.groupby(['column', 'column1']).[['column2', 'column3']].mean() # 다중 컬럼에 대한 결과 도출, 무조건 fancy indexing으로 가져와야함

df.groupby(['column', 'column1']).[['column2', 'column3']].agg(['mean', 'sum']) # 다중 통계 함수 적용, numpy도 적용 가능
```

```python
# pivot_table()
# groupby()와 동작이 유사하다
# index, columns, values를 지정하여 피벗한다
# groupby는 인덱스로 pivot_table만드는 것, pivot_table은 인덱스를 꺼내서 행과 열로 펼칠 수 있다는 게 가장 큰 차이점

df.pivot_table(index='column', values='column1') # index에 그룹을 표기
df.pivot_table(columns='column', values='column1') # columns에 그룹을 표기

df.pivot_table(index=['column', 'column2'], values='column1') # 다중 그룹에 대한 단일 컬럼 결과

df.pivot_table(index='column', columns='column1', values='column2') # 상단엔 1, 인덱스로는 0, 기본 값으로는 평균
df.pivot_table(index='column', columns='column1', values='column2', aggfunc=['','']) # 다중 통계함수 적용
```
